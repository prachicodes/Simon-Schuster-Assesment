# FROMDATE

```python
#converting the fromdate to year
data['fromdate']= data['fromdate'].astype(str).str[:4]
data.head()
```

```python
#here we transform the categorical to numeric 
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
data["formattype"] = lb_make.fit_transform(data["formattype"])
data.head()
```

HANDLING MISSING DATA

# AMAZONTOTALREVIEWS

```python
#HANDLING DATA FOR AMAZONTOTALREVIEWS 


#this fill the misisng value for Walking dead and formattype=1 with mean of amazontotalreview
m = data[(data['entity']=='The Walking Dead')  & (data['formattype'] == 1)]

#there are null values in amazontotalreviews hence filling it
m['amazontotalreviews'].fillna((m['amazontotalreviews'].mean()), inplace=True)   #filling null values with mean

#this fill the misisng value for Alone in the dark and formattype=1 with mean of amazontotalreview
m1 = data[(data['entity']=='Alone in the dark')  & (data['formattype'] == 1)]

#there are null values in amazontotalreviews hence filling it
data['amazontotalreviews'].fillna((m1['amazontotalreviews'].mean()), inplace=True)   #filling null values with mean
```

```python
m.info()
```

```python
m1.info()
```

```python
frames=[m1,m]
data=pd.concat(frames)
data
```
